## **Importing required libraries**

## **Dropping all unrequired fields to fetch necessary data for further processing (Preprocessing of fetched data).**

```python
# Data cleaning
df= pd.read_csv("../Data/Initialdataset.csv")
df2=df.drop(['ParentId','DeletionDate','OwnerDisplayName','OwnerUserId','FavoriteCount','CommunityOwnedDate','ContentLicense','ClosedDate','LastEditorUserId','LastEditorDisplayName','LastEditDate','LastActivityDate'],axis=1)
Postdf = df2.dropna().reset_index(drop=True)
Postdf
```

## **Adding respective Question ID to construct links and append to dataset.**

```python
# adding question link 
Postdf['link']= 'https://stackoverflow.com/q/'+ Postdf['Id'].astype(str)
os.makedirs('folder/subfolder', exist_ok=True)  
Postdf.to_csv('../Data/final_out.csv')  
```
